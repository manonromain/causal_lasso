### Solving using CVXPY in one step

We know that we have $Y = YB + \epsilon$ where $Y$ is a row observation, $B$ is strictly upper triangular and $\epsilon \sim^{iid} \mathcal{N}(0, \Omega)$. 

We have access to $X = YP^T$, so $X = (XPB + \epsilon)P^T$. So if $\Sigma = E(X^TX)$, $(I-PB^TP^T)\Sigma(I-PBP^T) = P\Omega P^T$, so $\Theta = \Sigma^{-1} = (I-PBP^T)P\Omega^{-1} P^T(I-PB^TP^T) = P(I-B)\Omega^{-1}(I-B^T)P^T$, let $U = \Omega^{-1/2}(I-B)$.

```python
def compare_graphs(A_gt, A, verbose=True):
    thres_A = np.abs(A) > 0.5
    undirected_A = thres_A + thres_A.T
    if verbose:
        print("Number of edges in gt", np.sum(A_gt))
        print("Number of edges in estimation", np.sum(thres_A))
    correct = np.sum(np.multiply(A_gt, thres_A))
    rev = np.sum(np.multiply(A_gt, np.multiply((1 - thres_A), thres_A.T)))
    missing = np.sum(A_gt) - np.sum(np.multiply(A_gt, undirected_A))
    extra = np.sum(thres_A) - np.sum(np.multiply(thres_A, (A_gt + A_gt.T)))
    shd = extra + rev + missing
    if verbose:
        print("Correct edges", correct)
        print("Reversed edges", rev)
        print("Missing edges", missing)
        print("Extra edges", extra)
        print("SHD", shd)
    return correct, extra, rev, missing, shd

print("Stats with only pos coeff")
correct, extra, rev, missing, shd = compare_graphs(adj_matrix, A)
print("\nStats with both pos and neg coeff")
correct, extra, rev, missing, shd = compare_graphs(adj_matrix, np.abs(A_m)) #np.reshape(logging_m["support"][14], (n,n))

# if False:
#     with open("results_pos_16_09.csv", "at") as f:
#         writer = csv.writer(f)
#         writer.writerow(["bregman_mosek", graph_type, n, mean_deg, noise_type, empirical_density,
#                          L1PENALTY, correct, extra, rev, missing, shd,
#                          16*60 +39])
```
